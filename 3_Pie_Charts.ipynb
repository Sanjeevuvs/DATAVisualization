```python
from matplotlib import pyplot as plt
```

<h3 style = "color:red"> Recommended to use when we have 5 or less values to plot </h3>

### 1. Plotting the Pie Chart

```python
slices = [30, 40, 20, 50] #sum needs not be 100
```

```python
plt.pie(slices)
plt.show()
```

### 2. Adding labels to the pie chart

```python
labels = ['thirty','fourty', 'twenty','fifty']
plt.pie(slices, labels=labels)
plt.show()
```

### 3. setting edge color

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'Blue'})
plt.show()
```

### 4. setting color of the slices

```python
color = ['blue','red','yellow','green']
 
#hexadecimal color codes can also be used
plt.pie(slices, labels=labels, colors=color, wedgeprops={'edgecolor':'black'})
plt.show()
```

### 5. plotting real world data

```python
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
 
slices = [59219, 55466, 47544, 36443, 35917]
```

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})
plt.show()
```

### 6. Exploding the slice

```python
explode=[0.3,0,0.8,0,0]
 
plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor':'black'})
plt.show()
```

### 7. adding shadow to the chart

```python
plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={'edgecolor':'black'})
plt.show()
```

### 8. setting the starting angle

```python
plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, wedgeprops={'edgecolor':'black'})
plt.show()
```

### 9. displaying percentage of each slices

```python
plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=60, autopct="%0.1f%%", wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/My Drive/data/Covid_19.csv')
```

```python
data.head()
```

```python
 slices=list(data["population"])
State=list(data["State"])
plt.pie(slices,shadow=True, startangle=60,wedgeprops={'edgecolor':'blue'})
plt.show()
```

```python
 
slices=list(data["population"])
State=list(data["State"])
textprops = {"fontsize":30}
plt.pie(slices,labels=State,shadow=True,pctdistance=1.1,labeldistance=1.2,radius=4,startangle=60, autopct="%0.1f%%",rotatelabels = 135,textprops = textprops,wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
slices =list(data["total__confirmed"])
textprops = {"fontsize":30}
plt.pie(slices,labels=State,shadow=True,pctdistance=1.3,labeldistance=1.6,radius=4,startangle=60, autopct="%0.1f%%",rotatelabels = 135,textprops = textprops,wedgeprops={'edgecolor':'black'})
plt.legend(loc='righit')
plt.show()
```

```python
 
 
slices=list(data["population"].head())
State=list(data["State"].head())
textprops = {"fontsize":10}
plt.pie(slices,labels=State,shadow=True,labeldistance=1,radius=1,startangle=60, autopct="%0.1f%%",rotatelabels = 135,textprops = textprops,wedgeprops={'edgecolor':'black'})
plt.legend(loc="lower left")
plt.title("The first 5 states and the population of the states from the covid19")
plt.show()
```

```python
slices=list(data["total__confirmed"].head())
State=list(data["State"].head())
textprops = {"fontsize":10}
plt.pie(slices,labels=State,shadow=True,labeldistance=1,radius=1,startangle=60, autopct="%0.1f%%",rotatelabels = 135,textprops = textprops,wedgeprops={'edgecolor':'blue'})
plt.legend(loc="lower left")
plt.title("Thefirst 5 states with total_confirmed cases covid19")
plt.show()
```

```python
 
slices=list(data["total__deceased"].head(6))
State=list(data["State"].head(6))
textprops = {"fontsize":10}
explode=[0.35,0,0,0.3,0.4,0.1]
plt.pie(slices,labels=State,explode=explode,shadow=True,labeldistance=1,radius=1,startangle=200, autopct="%0.1f%%",rotatelabels = 135,textprops = textprops,wedgeprops={'edgecolor':'blue'})
plt.legend(loc="best")
plt.title("Thefirst 5 states with total__deceased cases covid19")
plt.show()
```

```python
 
 
slices=list(data["total__recovered"].tail(6))
State=list(data["State"].tail(6))
textprops = {"fontsize":10}
explode=[0.1,0,0,0.05,0.2,0.02]
plt.pie(slices,labels=State,explode=explode,shadow=True,labeldistance=1,radius=1,startangle=225, autopct="%0.1f%%",rotatelabels = 135,textprops = textprops,wedgeprops={'edgecolor':'blue'})
plt.legend(loc="upper right")
plt.title("The tail 5 states with total__recovered cases covid19")
plt.show()
```

```python
 
 
 
slices=list(data["total__tested"].tail(6))
State=list(data["State"].tail(6))
textprops = {"fontsize":10}
explode=[0.1,0,0,0.05,0.2,0.02]
plt.pie(slices,labels=State,explode=explode,shadow=True,labeldistance=1,radius=1,startangle=90, autopct="%0.1f%%",rotatelabels = 135,textprops = textprops,wedgeprops={'edgecolor':'blue'})
plt.legend(loc="upper left")
plt.title("The tail 5 states with total__tested cases covid19")
plt.show()
```
