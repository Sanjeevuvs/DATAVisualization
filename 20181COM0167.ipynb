**Q4. Create the below directed graph(G)** *(5 Marks)*


Q4_image.png

```
#answer here
g1 = Digraph('G1', filename='G1.gv')

g1.node('Start')
g1.node('Hello')
g1.node('World')
g1.node('Data')
g1.node('Visualization')
g1.node('Stop')

g1.edge('Start','Hello')
g1.edge('Start','Data')

g1.edge('Hello','World')
g1.edge('Data','Visualization')

g1.edge('World','Stop')
g1.edge('Visualization','Stop')

g1.view()

```
