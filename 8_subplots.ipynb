```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv("/content/drive/My Drive/data/subplot_data.csv")
```

```python
data.head()
```

```python
ages = data['Age']
 
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
fig, ax = plt.subplots()
 
ax.plot(ages, dev_salaries, label="All Developers")
ax.plot(ages, py_salaries, label = "Python Developers")
ax.plot(ages, js_salaries, linestyle = "--", label = "Javascript Developers")
 
ax.legend()
ax.set_title("Developers salaries with age")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
 
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1)
 
ax1.plot(ages, dev_salaries, label="All Developers")
ax2.plot(ages, py_salaries, label = "Python Developers")
ax2.plot(ages, js_salaries, linestyle = "--", label = "Javascript Developers")
 
ax1.legend()
ax1.set_title("Developers salaries with age")
ax1.set_ylabel("Salary")
 
ax2.legend()
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary")
 
plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)
 
ax1.plot(ages, dev_salaries, label="All Developers")
ax2.plot(ages, py_salaries, label = "Python Developers")
ax2.plot(ages, js_salaries, linestyle = "--", label = "Javascript Developers")
 
ax1.legend()
ax1.set_title("Developers salaries with age")
ax1.set_ylabel("Salary")
 
ax2.legend()
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary")
 
plt.tight_layout()
plt.show()
```

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
 
ax1.plot(ages, dev_salaries, label="All Developers")
ax2.plot(ages, py_salaries, label = "Python Developers")
ax2.plot(ages, js_salaries, linestyle = "--", label = "Javascript Developers")
 
ax1.legend()
ax1.set_title("Developers salaries with age")
ax1.set_ylabel("Salary")
 
ax2.legend()
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary")
 
plt.tight_layout()
plt.show()
```
