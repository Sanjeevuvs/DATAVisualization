```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/My Drive/data/salary_data-1.csv")
```

```python
data.head()
```

```python
ages = data["Age"]
 
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]
```

### 1. Line Plot

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
plt.legend()
plt.show()
```

### 2. Adding Plot fill to line plot

```python
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.fill_between.html
```

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
plt.legend()
plt.fill_between(ages, py_salaries)
plt.show()
```

### 3. setting alpha value

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
plt.legend()
plt.fill_between(ages, py_salaries, alpha =0.35)
plt.show()
```

### 4. setting y2 value

```python
overall_median = 57287
```

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
plt.legend()
plt.fill_between(ages, py_salaries,overall_median, alpha = 0.25)
plt.show()
```

### 5. condition on plot fill

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
plt.legend()
plt.fill_between(ages, py_salaries,overall_median, where=(py_salaries > overall_median), interpolate=True, alpha = 0.25)
plt.show()
```

### 6. different plot fills in the same plot

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
plt.legend()
plt.fill_between(ages, py_salaries,overall_median, where=(py_salaries > overall_median), interpolate=True, alpha = 0.25)
plt.fill_between(ages, py_salaries,overall_median, where=(py_salaries <= overall_median), interpolate=True, alpha = 0.25)
plt.show()
```

### 7. changing plot fill color

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
plt.legend()
plt.fill_between(ages, py_salaries,overall_median, where=(py_salaries > overall_median), interpolate=True, alpha = 0.25)
plt.fill_between(ages, py_salaries,overall_median, where=(py_salaries <= overall_median), interpolate=True, color="red", alpha = 0.25)
plt.show()
```

### 8. Filling the area between two plots

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
plt.legend()
 
plt.fill_between(ages, py_salaries,dev_salaries, 
                 where=(py_salaries > dev_salaries), 
                 interpolate=True, 
                 alpha = 0.25)
 
plt.fill_between(ages, py_salaries,dev_salaries, 
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, color="red",
                 alpha = 0.25)
plt.show()
```

### 9. adding legend to the plot fills

```python
plt.plot(ages, dev_salaries, linestyle = "--",label="All developers")
plt.plot(ages, py_salaries, label = "Python Developers")
plt.xlabel("Age")
plt.ylabel("Salaries in USD")
plt.title("Developer Salaries in USD with Age")
 
plt.fill_between(ages, py_salaries,dev_salaries, 
                 where=(py_salaries > dev_salaries), 
                 interpolate=True,label="Above Average", 
                 alpha = 0.25)
 
plt.fill_between(ages, py_salaries,dev_salaries, 
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, color="red",label="Below Average", 
                 alpha = 0.25)
plt.legend(loc="lower right")
plt.show()
```
